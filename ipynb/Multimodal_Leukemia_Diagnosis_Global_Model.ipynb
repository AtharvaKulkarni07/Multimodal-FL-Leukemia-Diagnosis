```python
import socket
import subprocess
import time
import flwr as fl
import nest_asyncio
import urllib.request

nest_asyncio.apply()
print("Libraries installed.")
```

### Finding Free Port for Communication

```python
def find_free_port():
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind(('', 0))
        return s.getsockname()[1]

FREE_PORT = find_free_port()
print(f"Found an open door at Port: {FREE_PORT}")
```

### Connecting to the port


```python
with open("url.txt", "w") as f:
    subprocess.Popen(["lt", "--port", str(FREE_PORT)], stdout=f)

time.sleep(5)
with open("url.txt", "r") as f:
    print(f"PUBLIC URL: {f.read()}")
```

```python
nest_asyncio.apply()

external_ip = urllib.request.urlopen('https://ident.me').read().decode('utf8')
print(f"Use this IP to verify the link: {external_ip}")
```

### FedAvg

```python
strategy = fl.server.strategy.FedAvg(
    min_fit_clients=2,
    min_available_clients=2,
    min_evaluate_clients=2,
)

print(f"Server starting on port {FREE_PORT}...")
```

```python
try:
    fl.server.start_server(
        server_address=f"0.0.0.0:{FREE_PORT}",
        config=fl.server.ServerConfig(num_rounds=3),
        strategy=strategy
    )
except Exception as e:
    print(f"ERROR: {e}")
```
